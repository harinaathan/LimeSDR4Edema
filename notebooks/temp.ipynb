```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
baseband_freq = 100e3
baseband_sample_count = 1024
baseband_time = (np.arange(baseband_sample_count) / 2e6)
tx_samples = np.sin(baseband_time * 2 * np.pi * baseband_freq)

carrier_freq = 2.4e9
carrier_time = (np.arange(baseband_sample_count) / carrier_freq)
carrier_samples = np.sin(carrier_time * 2 * np.pi * carrier_freq)


plt.plot(baseband_time, tx_samples, label = 'baseband', color = 'r')
plt.plot(carrier_time, tx_samples, label = 'carrier', color = 'b')
plt.xlim((0,0.000012))
plt.show()

plt.plot(baseband_time, tx_samples, label = 'baseband', color = 'r')
plt.plot(carrier_time, tx_samples, label = 'carrier', color = 'b')
plt.xlim((0,0.0000005))
plt.show()
```
